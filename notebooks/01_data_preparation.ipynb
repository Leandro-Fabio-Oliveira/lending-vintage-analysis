# 01 — Data Preparation

The LendingClub dataset contains ~2.2 million loan records from 2007 to 2018. Before any analysis, we need to understand the structure, quality, and limitations of the data.

This notebook covers:
- Dataset structure and column overview
- Identifying columns relevant to vintage analysis
- Missing values and low variance assessment
- Duplicated rows

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
%%time
file = r'..\data\raw\accepted_2007_to_2018q4\accepted_2007_to_2018Q4.csv'
df_raw = pd.read_csv(file)
;
```

```python
df_raw.head(3)
```

```python
df_raw.shape
```

As the Dataset has 151 columns, lets take a look if there are columns completely missing, or filled with just a unique value. If we find it, we can remove before any exploration.

```python
%%time
df_nunique = (
    df_raw
    .nunique()
    .rename_axis('column')
    .to_frame('n_unique')
)

df_nunique.head(10)
```

```python
%%time
df_missing = (
    df_raw
    .isna()
    .sum()
    .rename('n_missing')
    .rename_axis('column')
    .to_frame()
)

df_missing['pct'] = df_missing['n_missing'] / len(df_raw)

df_missing.head(10)
```

```python
# Columns with more than 70% missing
high_missing = df_missing.loc[df_missing['pct'] > 0.70].index.tolist()

# Columns with 0 or 1 unique value (no analytical value)
low_variance = df_nunique.loc[df_nunique['n_unique'] <= 1].index.tolist()

# Columns irrelevant by definition (admin/id fields)
admin_cols = ['member_id', 'desc', 'url', 'id']

columns_to_remove = list(set(high_missing + low_variance + admin_cols))

print(f'Columns to remove: {len(columns_to_remove)}')
print(f'Remaining columns: {df_raw.shape[1] - len(columns_to_remove)}')
```

```python
df = df_raw.drop(columns=columns_to_remove).copy()
print(f'Shape after removal: {df.shape}')
```

After removing columns with more than 70% missing and those with low variance, the dataset reduced from 151 to 107 columns. However, 107 columns is still a large number to analyse. Fortunately, the dataset's Kaggle's page includes a link to the data dictionary ([see here](https://data.world/jaypeedevlin/lending-club-loan-data-2007-11)). We can use it to identify additional columns that may be unnecessary for our analysis.

For this project, the variables will be separated into three groups: Characteristics related to the loan origination, outcomes after origination, and Bureau information.

```python
data_dict_raw = pd.read_csv(r'..\data\raw\LCDataDictionary.csv')
data_dict = data_dict_raw.loc[data_dict_raw['LoanStatNew'].isin(df.columns.tolist())]
```

```python
cols_selected = [
    # --- Core vintage ---
    'issue_d',                # The month which the loan was funded
    'loan_status',            # outcome variable
    'loan_amnt',              # loan amount
    'grade',                  # risk grade (A to G)
    'sub_grade',              # granular risk grade
    'term',                   # 36 or 60 months
    'int_rate',               # interest rate
    'installment',            # monthly payment
    
    # --- At-origination borrower profile ---
    'annual_inc',             # self-reported income
    'dti',                    # debt-to-income ratio
    'emp_length',             # employment length
    'home_ownership',         # RENT / OWN / MORTGAGE / OTHER
    'addr_state',             # the state provided by the borrower in the loan application
    'purpose',                # loan purpose
    'verification_status',    # indicates if income was verified or not
    'application_type',       # individual or joint loan
    'fico_range_low',         # FICO score at origination
    'fico_range_high',        # FICO score at origination
    'earliest_cr_line',       # credit history length
    'delinq_2yrs',            # past delinquency
    'inq_last_6mths',         # recent credit inquiries
    'open_acc',               # open credit lines
    'pub_rec',                # derogatory records
    'revol_util',             # revolving utilization
    'mths_since_last_delinq', # months since last delinquency
    
    # --- Post-origination outcomes ---
    'last_pymnt_d',           # last month payment was received
    'last_pymnt_amnt',        # last total payment amount received
    'total_pymnt',            # total received to date
    'total_rec_prncp',        # principal received to date
    'total_rec_late_fee',     # Late fees received to date
    'recoveries',             # post charge off gross recovery
    'out_prncp',              # outstanding principal
]

df = df[cols_selected].copy()
print(f'Final shape: {df.shape}')
```

Now that we have selected our columns, let's take a look to see if there are duplicated rows or rows with all missing value.

```python
print(f'Number of duplicated rows: {df.duplicated().sum()}')
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(f'Rows with all missing values: {df.isna().all(axis=1).sum()}')
```

```python
df.dropna(how='all', inplace=True)
```

With our data preparation complete, we'll save our dataset for the next step: Exploratory Data Analysis.
